# Estimate the value of PI using Monte Carlo

Imagine a quadrant of a circle is placed inside a square, as shown next, and
we generate some random points inside the square. You can see that some of the points fall
inside the circle while others are outside the circle:





```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Define the Parameters

```python
square_size = 1
number_of_points = 100000
arc = np.linspace(0, np.pi / 2, 100)
```

### Generate the Random Points

```python
points = np.random.rand(number_of_points, 2)
```

### Visualization

```python
plt.figure(figsize=(5, 5))
plt.scatter(points[:, 0], points[:, 1], s=4)
plt.plot( square_size * np.cos(arc), square_size * np.sin(arc), c='r')
plt.show()
```

### Calculate PI

```python
number_of_points_inside_circle = np.sum(points[:, 0] ** 2 + points[:, 1] ** 2 <= square_size)
```

```python
pi = 4 * (number_of_points_inside_circle / number_of_points)
print(pi)
```
